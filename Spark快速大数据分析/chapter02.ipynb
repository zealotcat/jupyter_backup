## 2.3 Spark核心概念简介
从上层来看，每个 Spark 应用都由一个驱动器程序（driver program）来发起集群上的各种并行操作。驱动器程序包含应用的 main 函数，并且定义了集群上的分布式数据集，还对这些分布式数据集应用了相关操作。当使用Spark Shell时，实际的驱动器程序就是 Spark shell 本身，你只需要输入想要运行的操作就可以了。

驱动器程序通过一个 SparkContext 对象来访问 Spark。这个对象代表对计算集群的一个连接。shell 启动时已经自动创建了一个 SparkContext 对象，是一个叫作 sc 的变量。

```python
>>> sc
<SparkContext master=local[*] appName=PySparkShell>
```

一旦有了 SparkContext，你就可以用它来创建 RDD，比如读取一个文本文件之后就创建了一个RDD；之后可以在这些RDD上进行各种操作，比如用count()统计文件的行数。

要执行这些操作，驱动器程序一般要管理多个执行器（executor）节点。比如，如果我们在集群上运行 count() 操作，那么不同的节点会统计文件的不同部分的行数。下图展示了 Spark 如何在一个集群上运行。
